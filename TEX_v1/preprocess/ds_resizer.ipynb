```python
from PIL import Image
import os
import PIL
import glob
```

```python
path = '../ds/TEA/'

dirs = os.listdir(path)
new_path = '../intel/resized_imgs'
os.mkdir(new_path)
```

```python
for el in dirs:
    os.mkdir(new_path +'/' + el)
    for i, img in enumerate(os.listdir(path + el)):
        image = Image.open(path + el + '/' + img)
        print(image.size)
        new_img = image.resize((150, 150), PIL.Image.NEAREST)
        new_img.save(new_path + '/' + el +'/' + str(el) + '_' + str(i) + '.jpg')
```

```python
test_path = '../ds/tes_imgs/'
new_test_path = '../intel/resized_test_imgs'
dirs = os.listdir(test_path)
os.mkdir(new_test_path)

for i, el in enumerate(dirs):
    print(el)
    image = Image.open(test_path + el)
    print(image.size)
    new_img = image.resize((150, 150), PIL.Image.NEAREST)
    new_img.save(new_test_path + '/' + str(el) + '_' + str(i) + '.jpg')
```
